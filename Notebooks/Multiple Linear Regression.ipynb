```python
def initialize(df):
    
    x1 = df.iloc[:,[2]] #independent variables
    x1 = np.array(x1)
    x2 = df.iloc[:,[3]] #independent variables
    x2 = np.array(x2)
    y = df.iloc[:,[1]] #dependent variable
    y = np.array(y)
    m1 = 1
    m2 = 1
    c = 1
    
    return x1,x2,y,m1,m2,c
```

```python
def train(x1,x2,y,m1,m2,c,itr,lr):
    
    for i in range(itr):
        
        # y = m1x1 +m2x2 + c
        y1 = (m1*x1) + (m2*x2) + c
            
        dm1 = np.mean((y1-y)*x1) # gradient of slope m1
        dm2 = np.mean((y1-y)*x2) # gradient of slope m2
        dc = np.mean((y1-y)) # gradient of intercept c 
        
        m1 = m1 - lr*dm1 # update slope m1
        m2 = m2 - lr*dm2 # update slope m1
        c = c -lr*dc # update intercept c 
        
        L = np.mean(((y1-y))**2) # MSE Loss function
        print("Loss: ", L)
        
    y1 = m1*x1 + m2*x2 + c
    
    fig = plt.figure(figsize=(20,10))
    ### Set figure size
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x1,x2.T,y,c='red', marker='o', alpha=0.5)
    ax.plot_surface(x1,x2.T,y1, cmap ='viridis', alpha=0.05)
    plt.show()
    
    
    return m1,m2,c
        
```

```python
x1,x2,y,m1,m2,c = initialize(df)
```

```python
m1,m2,c = train(x1,x2,y,m1,m2,c,itr=10,lr=0.01)
```
